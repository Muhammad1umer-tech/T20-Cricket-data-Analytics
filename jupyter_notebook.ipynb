# Importing Libraries

```python
import json
import csv 
import pandas as pd
```

# Loading batting json file and converting it into csv after transformation

```python

json_file_path = 't20_json_files/t20_wc_batting_summary.json'

with open(json_file_path) as json_file:
    data = json.load(json_file)
```

```python
data[0]['battingSummary'][0:2]
```

```python
d = data[0]['battingSummary'][0].keys()
field_names = list(d)
```

```python
batting_summary_list = data
csv_file_name = "test/bating_summary.csv"
with open(csv_file_name, mode='w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=field_names)
    
    writer.writeheader()
    
    for batting_summary in batting_summary_list:
        for val in batting_summary['battingSummary']:
            writer.writerow(val)

print(f"The data has been saved to {csv_file_name}")
```

```python
df_batting = pd.read_csv("test/bating_summary.csv")
df_batting['dismissal'] = df_batting['dismissal'].apply(lambda x: 'not out' if pd.isna(x) or x.strip() == '' else 'out')
df_batting.to_csv("test/bating_summary.csv");
df_batting = pd.read_csv("test/bating_summary.csv")

df_batting
```

# Loading player_info json file and converting it into csv after transformation

```python
json_file_path = 't20_json_files/t20_wc_player_info.json'

with open(json_file_path) as json_file:
    data = json.load(json_file)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players.to_csv("test/dim_players.csv");
df_players
```
